**STEP 3 : EXPLORATORY DATA ANALYSIS** 

```python
df.columns
```

**Gender**

```python
sns.countplot(data=df,x='Gender',hue='Gender', palette='deep')
plt.title("Count plot for Gender")
plt.show()
```

```python
count=sns.countplot(data=df,x='Gender',hue='Gender', palette='muted')
for bars in count.containers:
    count.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(data=sales_gen,x='Gender',y='Amount')
```

**Form this above graph we can see that most of the buyers are females and even the purchasing power of the female are gratter than men !!**

**Age**

```python
df.columns
```

```python
ax=sns.countplot(x='Age Group', hue='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_age,x='Age Group',y='Amount',hue='Age Group',palette='bright')
```

**From the above graph we can see that most of buyers are in the age group between 26 to 35 years female**

**State**

```python
plt.figure(figsize=(15,5))
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=sales_state,x='State',y='Amount',hue='State',palette='dark')
```

```python
plt.figure(figsize=(12,5))
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=sales_state,x='State',y='Orders',hue='State',palette='Set3')
```

**Form the above graph we can see that most of the orders from the Uttar pradesh ,Maharashtra and Karnataka**

**Marital_Status**

```python
plt.figure(figsize=(5,5))
ms=sns.countplot(data=df,x='Marital_Status',hue='Marital_Status',palette='Set2')
for bar in ms.containers:
    ms.bar_label(bar)
```

```python
plt.figure(figsize=(8,5))
sales_ms=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_ms,x='Marital_Status',y='Amount',hue='Gender')
```

**From the above graph we can see that most of buyers are married(women) and they have high purchasing power**

**Occupation**

```python
plt.figure(figsize=(15,5))
oc=sns.countplot(data=df,x='Occupation', hue='Occupation',palette='Set2')
for bar in oc.containers:
    oc.bar_label(bar)
```

```python
plt.figure(figsize=(15,5))
sales_ms=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_ms,x='Occupation',y='Amount', hue='Occupation',palette='husl')
```

**From this above graph we can see tha most of the buyers are working in IT Sector,Healthcare and Aviation**

**Product_Catagory**

```python
plt.figure(figsize=(15,5))
oc=sns.countplot(data=df,x='Product_Category', hue='Product_Category',palette='Set2')
for bar in oc.containers:
    oc.bar_label(bar)
```

```python
plt.figure(figsize=(25,5))
sales_pc=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_pc,x='Product_Category',y='Amount', hue='Product_Category',palette='husl')
```

**From this above graph we can see that most of the sold products are food,clothing and footware**

**Product_ID**

```python
plt.figure(figsize=(15,5))
sales_pid=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=sales_pid,x='Product_ID',y='Orders', hue='Product_ID',palette='husl')
```

```python
fig1,ax1=plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

***Conclution***

**Married women age group 26-35 years from UP,Maharastra and Karnataka working in IT,Healthcare and Aviation sector are more likely to buy products from food, clothing and electronics category**
